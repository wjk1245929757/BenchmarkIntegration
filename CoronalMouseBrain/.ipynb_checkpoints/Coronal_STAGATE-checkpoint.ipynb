```python
datasets = ['DAPI', 'FFPE']
# datasets = ['DAPI', 'FreshFrozen']
# datasets = ['FFPE', 'FreshFrozen']
# datasets = ['DAPI', 'FFPE', 'FreshFrozen']

expriment_name = 'DAPI_FFPE_'
input_dir = 'G:/dataset/05-CoronalMouseBrain/input/'  # Replace it with your file path
output_dir = 'G:/dataset/05-CoronalMouseBrain/output/STAGATE/'  # Replace it with your file path

rad_cutoff_dict = {'DAPI': 300, 'FFPE': 300, 'FreshFrozen': 150}  # neighbor 5-10

# try    num_cluster = 12
num_cluster = 12
```

# load data

```python
adata_list = {}
remove_spot_batch_prefix_dict = {'DAPI': 'DAPI', 'FFPE': 'FFPE', 'FreshFrozen': 'Normal'}

for dataset in datasets:
    input_path = input_dir + dataset
    temp_adata = sc.read_visium(path=input_path, count_file='filtered_feature_bc_matrix.h5', load_images=True)
    temp_adata.var_names_make_unique()

    # Truth_df = pd.read_csv(os.path.join(input_dir, section_id+'_truth.txt'), sep='\t', header=None, index_col=0)
    # temp_adata.obs['Ground Truth'] = Truth_df.loc[temp_adata.obs_names, 1]

    # # read the annotation
    adata_label = pd.read_csv(os.path.join(input_path, 'truth.csv'), index_col=0)
    adata_label.index = [str.replace(str.replace(x, '.', '-'), remove_spot_batch_prefix_dict[dataset] + '-', '') for x in adata_label.index]
    temp_adata.obs['Ground Truth'] = adata_label.loc[temp_adata.obs_names, 'celltype'].astype('category')

    if dataset == 'FreshFrozen':
        adata.obsm['spatial'] = adata.obsm['spatial'].astype(np.int64)
    
    # make the spot name unique
    temp_adata.obs_names = [x+'_'+dataset for x in temp_adata.obs_names]

    adata_list[dataset] = temp_adata.copy()
```

```python
fig, axs = plt.subplots(1, len(datasets), figsize=(12, 3))
it=0
for dataset in datasets:
    if it == len(dataset):
        sc.pl.spatial(adata_list[dataset], img_key="hires", ax=axs[it],
                      color=["Ground Truth"], title=dataset, show=False)
    else:
        sc.pl.spatial(adata_list[dataset], img_key="hires", ax=axs[it], legend_loc=None,
                      color=["Ground Truth"], title=dataset, show=False)
    it+=1
```

```python
for dataset in datasets:
    STAGATE.Cal_Spatial_Net(adata_list[dataset], rad_cutoff=rad_cutoff_dict[dataset])
    STAGATE.Stats_Spatial_Net(adata_list[dataset])
```

# Conbat the scanpy objects and spatial networks

```python
adata = sc.concat([adata_list[x] for x in datasets], keys=None)
adata.uns['Spatial_Net'] = pd.concat([adata_list[x].uns['Spatial_Net'] for x in datasets])
STAGATE.Stats_Spatial_Net(adata)
```

# Calculate ARI

```python
from sklearn.metrics.cluster import adjusted_rand_score

for dataset in datasets:
    temp_adata = adata[adata.obs['batch_name']==dataset]
    temp_obs = temp_adata.obs.dropna()
    temp_ARI = adjusted_rand_score(temp_obs['mclust'], temp_obs['Ground Truth'])
    print('ARI of section ID %s: %.3f' %(dataset, temp_ARI))
```
