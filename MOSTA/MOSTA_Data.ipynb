```python
import os 
import scanpy as sc
import pandas as pd
import numpy as np
from anndata import AnnData
from pathlib import Path, PurePath
```

```python
data_name_list = ['E9.5_E1S1', 'E10.5_E2S1', 'E11.5_E1S1', 'E12.5_E1S1']
# data_path = '/home/lixiangyu/benchmark/data/MOSTA/'
# save_path = '/home/lixiangyu/benchmark/data/MOSTA/STAligner/'
# n_domains = 10

data_path = 'G:/dataset/08-MOSTA/input/'
```

```python
for dataset in data_name_list:
    adata = sc.read_h5ad(data_path + dataset + '.MOSTA.h5ad')
    print(adata)
    
    save_data_path = Path(os.path.join(data_path, dataset))
    print(save_data_path)
    save_data_path.mkdir(parents=True, exist_ok=True)

    count_exp = adata.layers['count'].todense()
    np.savetxt( os.path.join(save_data_path, 'count.csv'), count_exp, delimiter=',')
    data_exp = adata.X.todense()
    np.savetxt( os.path.join(save_data_path, 'data.csv'), count_exp, delimiter=',')

    cell_df = pd.DataFrame(data = adata.obs_names.tolist(),index = None,columns = None)
    cell_df.to_csv( os.path.join(save_data_path, 'cells.csv'), columns=None, header = False)

    gene_df = pd.DataFrame(data = list(adata.var_names),index = None,columns = None)
    gene_df.to_csv( os.path.join(save_data_path, 'genes.csv'), columns=None, header = False)

    np.savetxt( os.path.join(save_data_path, 'coord.csv'), adata.obsm['spatial'], delimiter=',')
```
