###  查看h5ad

```python
import scanpy as sc
import pandas as pd
import numpy as np

sample_names = list(range(151507, 151511))
sample_names.extend(list(range(151669, 151677)))
print(sample_names)

for sample_name in sample_names:
    # print(f"G:/dataset/1_DLPFC/output/SEDR/{sample_name}/{sample_name}.h5ad")
    adata = sc.read_h5ad(f"G:/dataset/1_DLPFC/output/SEDR/{sample_name}/{sample_name}.h5ad")
    # print(adata)
    np.savetxt(f"G:/dataset/1_DLPFC/output/SEDR/{sample_name}/{sample_name}_SEDR_embeddings.csv", adata.X, delimiter=",")

```
