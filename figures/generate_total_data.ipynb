# remove batch index

```python
batch_columns = ['PCR', 'iASW', 'iLISI', 'kBET', 'graph_connectivity']

batch_df = pd.DataFrame()
batch_df['method'] = df1['method']
batch_df['cluster_method'] = df1['cluster_method']
for column in batch_columns:
    if df_num == 1:
        batch_df[column] = df1[column]
    elif df_num == 2:
        batch_df[column] = (df1[column] + df2[column]) / 2
    elif df_num == 3:
        batch_df[column] = (df1[column] + df2[column] + df3[column]) / 3

batch_df = batch_df[batch_df['cluster_method'] == 'ground.truth' ]
batch_df['batch_mean'] = batch_df[batch_columns].mean(axis=1, numeric_only=True)
batch_df['batch_median'] = batch_df[batch_columns].median(axis=1, numeric_only=True)
batch_df = batch_df.sort_values(by='batch_mean',ascending=False)
batch_df
```

# save info

```python
info_columns = ['cASW', 'IsolatedLabelsAsw', 'cLISI', 'IsolatedLabelsF1', 'ARI', 'NMI']
info_df = pd.DataFrame()
info_df['method'] = df1['method']
info_df['cluster_method'] = df1['cluster_method']
for column in info_columns:
    if df_num == 1:
        info_df[column] = df1[column]
    elif df_num == 2:
        info_df[column] = (df1[column] + df2[column]) / 2
    elif df_num == 3:
        info_df[column] = (df1[column] + df2[column] + df3[column]) / 3

info_df
```

```python
info_df_ARI = info_df[['method', 'cluster_method', 'ARI']]
info_df_ARI = info_df_ARI[ info_df_ARI['cluster_method'] != 'ground.truth'  ]
info_df_ARI = info_df_ARI.sort_values(by='ARI',ascending=False)
print(info_df_ARI)

info_df_ARI = info_df_ARI.drop_duplicates(subset=['method'], keep='first')
print(info_df_ARI)
for index, row in info_df_ARI.iterrows():
    index_ = info_df.loc[ (info_df['cluster_method'] == 'ground.truth') & (info_df['method'] == row['method']) ].index.tolist()
    assert len(index_) == 1,'check shape'
    index_ = index_[0]
    info_df.loc[index_ , 'ARI'] = row['ARI']

info_df
```

```python
info_df_NMI = info_df[['method', 'cluster_method', 'NMI']]
info_df_NMI = info_df_NMI[ info_df_NMI['cluster_method'] != 'ground.truth'  ]
info_df_NMI = info_df_NMI.sort_values(by='NMI',ascending=False)

info_df_NMI = info_df_NMI.drop_duplicates(subset=['method'], keep='first')
info_df_NMI
for index, row in info_df_NMI.iterrows():
    index_ = info_df.loc[ (info_df['cluster_method'] == 'ground.truth') & (info_df['method'] == row['method']) ].index.tolist()
    assert len(index_) == 1,'check shape'
    index_ = index_[0]
    info_df.loc[index_ , 'NMI'] = row['NMI']

info_df
```

```python
info_df = info_df[info_df['cluster_method'] == 'ground.truth' ]
info_df['info_mean'] = info_df[info_columns].mean(axis=1, numeric_only=True)
info_df['info_median'] = info_df[info_columns].median(axis=1, numeric_only=True)
info_df = info_df.sort_values(by='info_mean',ascending=False)
info_df
```

# spatial index

```python

spatial_columns = ['spatial_coherence_score', "Moran's I", "Geary's C"]
spatial_df = pd.DataFrame()
spatial_df['method'] = df1['method']
spatial_df['cluster_method'] = df1['cluster_method']
for column in spatial_columns:
    if df_num == 1:
        spatial_df[column] = df1[column]
    elif df_num == 2:
        spatial_df[column] = (df1[column] + df2[column]) / 2
    elif df_num == 3:
        spatial_df[column] = (df1[column] + df2[column] + df3[column]) / 3
    
spatial_df["Geary's C"] = 1 - spatial_df["Geary's C"]
    
spatial_df
```

```python
spatial_df_scs = spatial_df[['method', 'cluster_method', 'spatial_coherence_score']]
spatial_df_scs = spatial_df_scs[ spatial_df_scs['cluster_method'] != 'ground.truth'  ]

print(spatial_df_scs)

spatial_df_scs['spatial_coherence_score'] = spatial_df_scs['spatial_coherence_score'].abs()
spatial_df_scs = spatial_df_scs.sort_values(by='spatial_coherence_score',ascending=False)
spatial_df_scs = spatial_df_scs.drop_duplicates(subset=['method'], keep='first')

print(spatial_df_scs)

max_value = spatial_df_scs['spatial_coherence_score'].max()
spatial_df_scs['spatial_coherence_score'] = spatial_df_scs['spatial_coherence_score'] / max_value
spatial_df_scs
for index, row in spatial_df_scs.iterrows():
    index_ = spatial_df.loc[ (spatial_df['cluster_method'] == 'ground.truth') & (spatial_df['method'] == row['method']) ].index.tolist()
    print(index_)
    assert len(index_) == 1,'check shape'
    spatial_df.loc[index_ , 'spatial_coherence_score'] = row['spatial_coherence_score']
```
