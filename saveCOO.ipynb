```python
import pandas as pd
import numpy as np
import os, sys
from scipy.sparse import coo_matrix
```

```python
import gc
# 'Fulv1', 'Fulv2', 'Veh1', 
samples = ['Veh2']

for sample in samples:
    print(sample)
    dir_path = 'G:/CancerDB/202106-liver cancer/LICA_GSE236681/'
    
    gene = []
    res = []
    
    coord = pd.read_csv(dir_path + sample + '/SV/LICA_' + sample + '_coord_sp.csv')    
    X = coord.loc[ : , "x"].tolist()
    Y = coord.loc[ : , "y"].tolist()
    
    chunksize = 3000
    for chunk in pd.read_csv(dir_path + sample + '/SV/LICA_' + sample + '_counts_sp.csv', chunksize=chunksize):       
        gene_tmp = chunk.iloc[:, 0].tolist()
        chunk.drop(['Unnamed: 0'], axis=1, inplace=True)
        coo = coo_matrix(chunk.values.T)

        for i,j,d in zip(coo.row, coo.col, coo.data):
            tmp = [gene_tmp[j], X[i], Y[i], d]
            res.append(tmp)
        gene += gene_tmp

    gene_df = pd.DataFrame(data = gene,index = None,columns = ['gene'])
    gene_df.to_csv(dir_path + sample + '/gene.csv')
    columns = ['gene', 'X', 'Y', 'expression']
    dt = pd.DataFrame(res, columns=columns)
    dt.to_csv(dir_path + sample + '/svg.csv', index=None, sep=',')
```
