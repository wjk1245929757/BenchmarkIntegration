```python
from pathlib import Path
import scanpy as sc
import pandas as pd
import numpy as np
from anndata import AnnData

from matplotlib.image import imread
import json


def load_data_merge(file_paths, samples):
    adata_list = []
    for file_path in file_paths:
        adata_temp = sc.read_h5ad(file_path)
        adata_temp.var_names_make_unique()
        adata_temp.obs_names_make_unique()
        adata_list.append(adata_temp)
    adata_pre = ad.concat(adata_list, index_unique="-", keys = samples)
    return adata_pre

def load_data(file_paths):
    adata_list = []
    for file_path in file_paths:
        adata_temp = sc.read_h5ad(file_path)
        adata_list.append(adata_temp)
    return adata_list

def add_visium_image(
    adata,
    path: str,
    library_id = None,
    load_images = True,
    source_image_path = None,
    quality ='hires',
) -> AnnData:
    
    path = Path(path)

    adata.uns["spatial"] = dict()

    from h5py import File

    if library_id is None:
        library_id = 'image'

    adata.uns["spatial"][library_id] = dict()

    if load_images:
        tissue_positions_file = (
            path / "spatial/tissue_positions.csv"
            if (path / "spatial/tissue_positions.csv").exists()
            else path / "spatial/tissue_positions_list.csv"
        )
        files = dict(
            tissue_positions_file=tissue_positions_file,
            scalefactors_json_file=path / "spatial/scalefactors_json.json",
            hires_image=path / "spatial/tissue_hires_image.png",
            lowres_image=path / "spatial/tissue_lowres_image.png",
        )

        # check if files exists, continue if images are missing
        for f in files.values():
            if not f.exists():
                if any(x in str(f) for x in ["hires_image", "lowres_image"]):
                    logg.warning(
                        f"You seem to be missing an image file.\n"
                        f"Could not find '{f}'."
                    )
                else:
                    raise OSError(f"Could not find '{f}'")

        adata.uns["spatial"][library_id]["images"] = dict()
        for res in ["hires", "lowres"]:
            try:
                adata.uns["spatial"][library_id]["images"][res] = imread(
                    str(files[f"{res}_image"])
                )
            except Exception:
                raise OSError(f"Could not find '{res}_image'")

        # read json scalefactors
        adata.uns["spatial"][library_id]["scalefactors"] = json.loads(
            files["scalefactors_json_file"].read_bytes()
        )

        # read coordinates
        positions = pd.read_csv(
            files["tissue_positions_file"],
            header=0 if tissue_positions_file.name == "tissue_positions.csv" else None,
            index_col=0,
        )
        positions.columns = [
            "in_tissue",
            "array_row",
            "array_col",
            "pxl_col_in_fullres",
            "pxl_row_in_fullres",
        ]

        adata.obs = adata.obs.join(positions, how="left")

        adata.obsm["spatial"] = adata.obs[
            ["pxl_row_in_fullres", "pxl_col_in_fullres"]
        ].to_numpy()
        adata.obs.drop(
            columns=["pxl_row_in_fullres", "pxl_col_in_fullres"],
            inplace=True,
        )

        # put image path in uns
        if source_image_path is not None:
            # get an absolute path
            source_image_path = str(Path(source_image_path).resolve())
            adata.uns["spatial"][library_id]["metadata"]["source_image_path"] = str(
                source_image_path
            )
            
    if library_id is None:
        library_id = list(adata.uns["spatial"].keys())[0]
    if quality == "fulres":
        image_coor = adata.obsm["spatial"]
        img = plt.imread(image_path, 0)
        adata.uns["spatial"][library_id]["images"]["fulres"] = img
    else:
        scale = adata.uns["spatial"][library_id]["scalefactors"][
            "tissue_" + quality + "_scalef"]
        image_coor = adata.obsm["spatial"] * scale
    adata.obs["imagecol"] = image_coor[:, 0]
    adata.obs["imagerow"] = image_coor[:, 1]
    adata.uns["spatial"][library_id]["use_quality"] = quality
    return adata
```

```python
samples = ['anterior1', 'anterior2', 'posterior1', 'posterior2']
adata_pre = load_data_merge(['G:/dataset/2_saggital/input/'+sample+'/'+sample+'.h5ad' for sample in samples], samples)
print(adata_pre)
my_methods = ['harmony']

# # seurat embedding
# seurat_embedding_pd = pd.read_csv('G:/dataset/2_saggital/output/Saggital_all_seurat_embedding.csv', index_col = 0)
# print(seurat_embedding_pd.shape)
# print(seurat_embedding_pd.iloc[: , :3].head())
# if adata_pre.shape[0] == seurat_embedding_pd.shape[0]:
#     adata_pre.obsm['seurat_embedding'] = seurat_embedding_pd.values
# else:
#     print('check shape')

# # harmony embedding
# harmony_embedding_pd = pd.read_csv('G:/dataset/2_saggital/output/Saggital_all_harmony_embedding.csv', index_col = 0)
# print(harmony_embedding_pd.shape)
# print(harmony_embedding_pd.iloc[: , :3].head())
# if adata_pre.shape[0] == harmony_embedding_pd.shape[0]:
#     adata_pre.obsm['harmony_embedding'] = harmony_embedding_pd.values
# else:
#     print('check shape')
# print(adata_pre)

# # PRECAST embedding
# PRECAST_embedding_pd = pd.read_csv('G:/dataset/2_saggital/output/Saggital_all_PRECAST_embedding.csv', index_col = 0)
# print(adata_pre)
# print(adata_pre.obs_names[:3])
# print(PRECAST_embedding_pd.iloc[: , :3].head())
# PRECAST_embedding_pd.rename(index=lambda s: s[:len(s)-1]+'-'+samples[int(s[-1])-1], inplace=True)
# print(PRECAST_embedding_pd.shape)
# print(PRECAST_embedding_pd.iloc[: , :3].head())
# # ######### 过滤细胞  取子集anndata
# adata_pre = adata_pre[PRECAST_embedding_pd.index.to_list(), :]
# if adata_pre.shape[0] == PRECAST_embedding_pd.shape[0]:
#     adata_pre.obsm['PRECAST_embedding'] = PRECAST_embedding_pd.values
# else:
#     print('check shape')
# print(adata_pre)


# # STAGATE embedding
# STAGATE_embedding_pd = pd.read_csv('G:/dataset/2_saggital/output/SpaceRanger1.0/STAGATE/all_STAGATE.csv', index_col = 0, header = None)
# print(STAGATE_embedding_pd.shape)
# print(STAGATE_embedding_pd.iloc[: , :3].head())
# assert adata_pre.shape[0] == STAGATE_embedding_pd.shape[0],"check shape"
# adata_pre.obsm['STAGATE_embedding'] = STAGATE_embedding_pd.values
# print(adata_pre)

# # SPIARL embedding
# SPIARL_embedding_pd = pd.read_csv('G:/dataset/2_saggital/output/SpaceRanger1.0/SPIRAL/SPIRAL_all_posterior1_all_posterior2_all_anterior1_all_anterior2_embed_256.csv', index_col = 0)
# print(SPIARL_embedding_pd.shape)
# print(SPIARL_embedding_pd.iloc[: , :3].head())
adata.obsm['SPIRAL_embedding'][:,4:].shape
# assert adata_pre.shape[0] == SPIARL_embedding_pd.shape[0],"check shape"
# adata_pre.obsm['SPIARL_embedding'] = SPIARL_embedding_pd.values
# print(adata_pre)

adatas = adata_pre
# cal neighbors
for method in my_methods:
    print(method)
    if 'GraphST' in method:
        sc.pp.neighbors(adatas, n_neighbors=50, use_rep= method+'_embedding', key_added = method+'_neighbors')
    else:
        sc.pp.neighbors(adatas, use_rep= method+'_embedding', key_added = method+'_neighbors')
    
print(adatas)

## no k    Best resolution: leiden 0.00  louvain 0.00
## k= 10:18,2   
## k= 10   Best resolution: leiden 0.17   louvain 0.25
## k= 12   Best resolution: leiden 0.23   louvain 0.35
## k= 14   Best resolution: leiden 0.25    louvain 0.42
## k= 16   Best resolution: leiden 0.35   louvain 0.46
## k= 18   Best resolution: leiden 0.44   louvain 0.63

for k in range(1):
    print('########################')
    print('k=', k)
    
    for method in my_methods:
        print(method)
        adatas.uns['neighbors'] = adatas.uns[method+'_neighbors'].copy()
        adatas.obsp['distances'] = adatas.obsp[method+'_neighbors_distances'].copy()
        adatas.obsp['connectivities'] = adatas.obsp[method+'_neighbors_connectivities'].copy()
        # leiden
        adatas = index.get_cluster_data(adatas, k, False, 'leiden', 'leiden_'+method, resolution = sorted(list(np.arange(0.1, 1, 0.01)), reverse=True))
        # louvain
        adatas = index.get_cluster_data(adatas, k, False, 'louvain', 'louvain_'+method, resolution = sorted(list(np.arange(0.1, 1, 0.01)), reverse=True))
        # mclust
        adatas = index.get_cluster_data(adatas, None, False, 'mclust', 'mclust_'+method, None, method+'_embedding')

    print(adatas)

    ##### spatial_coherence_score
    methods = []
    for method in my_methods:
        methods.append('leiden_'+method)
        methods.append('louvain_'+method)
        methods.append('mclust_'+method)

    for method in methods:
        print(method)
        knn = 6
        g, node_dict = index.generate_graph_from_labels(adatas, adatas.obs[method] ,knn)
        scs = index.spatial_coherence_score(g, node_dict, 100)
        adatas.uns['scs_' + method] = scs

    for method in methods:
        print(adatas.uns['scs_' + method])

    ##### Moran's I  
    for method in my_methods:
        print(method + ' Morans I :')
        adatas.uns['neighbors'] = adatas.uns[method+'_neighbors'].copy()
        adatas.obsp['distances'] = adatas.obsp[method+'_neighbors_distances'].copy()
        adatas.obsp['connectivities'] = adatas.obsp[method+'_neighbors_connectivities'].copy()
        moranI = np.mean(sc.metrics.morans_i(adatas, obsm = method+'_embedding'))
        adatas.uns['moranI_' + method] = moranI
        print(moranI)

    # for method in my_methods:
    #     print(method + ':')
    #     print(np.mean(sc.metrics.morans_i(adatas, obsm = method+'_embedding')))

    ##### Geary's C 
    for method in my_methods:
        print(method + ' Gearys C :')
        adatas.uns['neighbors'] = adatas.uns[method+'_neighbors'].copy()
        adatas.obsp['distances'] = adatas.obsp[method+'_neighbors_distances'].copy()
        adatas.obsp['connectivities'] = adatas.obsp[method+'_neighbors_connectivities'].copy()
        gearysC = np.mean(sc.metrics.gearys_c(adatas, obsm = method+'_embedding'))
        adatas.uns['gearysC_' + method] = gearysC
        print(gearysC)


    ##### ARI
    methods = ['celltype']
    for method in my_methods:
        methods.append('leiden_'+method)
        methods.append('louvain_'+method)
        methods.append('mclust_'+method)

    print('ARI:')
    for method in methods:
        adatas.uns['ARI_' + method] = index.ARI(adatas, method, 'celltype')
        print(adatas.uns['ARI_' + method])
        print(scib.metrics.ari(adatas, method, 'celltype'))

    ##### NMI
    print('NMI:')
    for method in methods:
        adatas.uns['NMI_' + method] = index.NMI(adatas, method, 'celltype')
        print(adatas.uns['NMI_' + method])

    ##### cASW
    print('cASW:')
    for method in my_methods:
        adatas.uns['cASW_celltype'+method] = index.cASW(adatas, 'celltype', method+'_embedding')
        adatas.uns['cASW_leiden_'+method] = index.cASW(adatas, 'celltype', method+'_embedding')
        adatas.uns['cASW_louvain_'+method] = index.cASW(adatas, 'celltype', method+'_embedding')
        adatas.uns['cASW_mclust_'+method] = index.cASW(adatas, 'celltype', method+'_embedding')
        print(adatas.uns['cASW_celltype'+method])
        print(adatas.uns['cASW_leiden_'+method])
        print(adatas.uns['cASW_louvain_'+method])
        print(adatas.uns['cASW_mclust_'+method])

    ##### cLISI
    print('cLISI:')
    for method in my_methods:
        adatas.uns['cLISI_celltype'+method] = index.cLISI(adatas, 'celltype', type_="embed", use_rep=method+'_embedding')
        adatas.uns['cLISI_leiden_'+method] = index.cLISI(adatas, 'leiden_'+method, type_="embed", use_rep=method+'_embedding')
        adatas.uns['cLISI_louvain_'+method] = index.cLISI(adatas, 'louvain_'+method, type_="embed", use_rep=method+'_embedding')
        adatas.uns['cLISI_mclust_'+method] = index.cLISI(adatas, 'mclust_'+method, type_="embed", use_rep=method+'_embedding')
        print(adatas.uns['cLISI_celltype'+method])
        print(adatas.uns['cLISI_leiden_'+method])
        print(adatas.uns['cLISI_leiden_'+method])
        print(adatas.uns['cLISI_leiden_'+method])

    ##### IsolatedLabelsAsw
    import gc
    print('IsolatedLabelsAsw:')
    for method in my_methods:
        print(method)
        adatas.uns['IsolatedLabelsAsw_celltype'+method] = index.IsolatedLabelsAsw(adatas, batch_key="batch", label_key="celltype", embed=method+"_embedding")
        gc.collect()
        adatas.uns['IsolatedLabelsAsw_leiden_'+method] = index.IsolatedLabelsAsw(adatas, batch_key="batch", label_key="leiden_"+method, embed=method+"_embedding")
        gc.collect()
        adatas.uns['IsolatedLabelsAsw_louvain_'+method] = index.IsolatedLabelsAsw(adatas, batch_key="batch", label_key="louvain_"+method, embed=method+"_embedding")
        gc.collect()
        adatas.uns['IsolatedLabelsAsw_mclust_'+method] = index.IsolatedLabelsAsw(adatas, batch_key="batch", label_key="mclust_"+method, embed=method+"_embedding")
        gc.collect()
        print(adatas.uns['IsolatedLabelsAsw_celltype'+method])
        print(adatas.uns['IsolatedLabelsAsw_leiden_'+method])
        print(adatas.uns['IsolatedLabelsAsw_louvain_'+method])
        print(adatas.uns['IsolatedLabelsAsw_mclust_'+method])

    ##### IsolatedLabelsF1
    print('IsolatedLabelsF1:')
    for method in my_methods:
        print(method)
        adatas.uns['IsolatedLabelsF1_celltype'+method] = index.IsolatedLabelsF1(adatas, batch_key="batch", label_key="celltype", embed=method+"_embedding")
        gc.collect()
        adatas.uns['IsolatedLabelsF1_leiden_'+method] = index.IsolatedLabelsF1(adatas, batch_key="batch", label_key="leiden_"+method, embed=method+"_embedding")
        gc.collect()
        adatas.uns['IsolatedLabelsF1_louvain_'+method] = index.IsolatedLabelsF1(adatas, batch_key="batch", label_key="louvain_"+method, embed=method+"_embedding")
        gc.collect()
        adatas.uns['IsolatedLabelsF1_mclust_'+method] = index.IsolatedLabelsF1(adatas, batch_key="batch", label_key="mclust_"+method, embed=method+"_embedding")
        gc.collect()
        print(adatas.uns['IsolatedLabelsF1_celltype'+method])
        print(adatas.uns['IsolatedLabelsF1_leiden_'+method])
        print(adatas.uns['IsolatedLabelsF1_louvain_'+method])
        print(adatas.uns['IsolatedLabelsF1_mclust_'+method])

    ##### iASW
    print('iASW:')
    for method in my_methods:
        print(method)
        adatas.uns['iASW_celltype'+method] = index.iASW(adatas, batch_key="batch", label_key="celltype", embed=method+"_embedding")
        adatas.uns['iASW_leiden_'+method] = index.iASW(adatas, batch_key="batch", label_key="leiden_"+method, embed=method+"_embedding")
        adatas.uns['iASW_louvain_'+method] = index.iASW(adatas, batch_key="batch", label_key="louvain_"+method, embed=method+"_embedding")
        adatas.uns['iASW_mclust_'+method] = index.iASW(adatas, batch_key="batch", label_key="mclust_"+method, embed=method+"_embedding")
        print(adatas.uns['iASW_celltype'+method])
        print(adatas.uns['iASW_leiden_'+method])
        print(adatas.uns['iASW_louvain_'+method])
        print(adatas.uns['iASW_mclust_'+method])

    ##### iLISI
    print('iLISI:')
    for method in my_methods:
        print(method)
        adatas.uns['iLISI_celltype'+method] = index.iLISI(adatas, "batch", type="embed", use_rep=method+'_embedding')
        print(adatas.uns['iLISI_celltype'+method])

    ##### graph_connectivity
    print('graph_connectivity:')
    for method in my_methods:
        print(method)
        adatas.uns['graph_connectivity_celltype'+method] = index.graph_connectivity(adatas, type_="embed", label_key='celltype')
        adatas.uns['graph_connectivity_leiden_'+method] = index.graph_connectivity(adatas, type_="embed", label_key='leiden_'+method)
        adatas.uns['graph_connectivity_louvain_'+method] = index.graph_connectivity(adatas, type_="embed", label_key='louvain_'+method)
        adatas.uns['graph_connectivity_mclust_'+method] = index.graph_connectivity(adatas, type_="embed", label_key='mclust_'+method)
        print(adatas.uns['graph_connectivity_celltype'+method])
        print(adatas.uns['graph_connectivity_leiden_'+method])
        print(adatas.uns['graph_connectivity_louvain_'+method])
        print(adatas.uns['graph_connectivity_mclust_'+method])

    ##### kBET
    print('kBET:')
    for method in my_methods:
        print(method)
        adatas.uns['kBET_celltype'+method] = index.kBET(adatas, batch_key="batch", label_key='celltype', type_="embed", embed=method+"_embedding")
        adatas.uns['kBET_leiden_'+method] = index.kBET(adatas, batch_key="batch", label_key='leiden_'+method, type_="embed", embed=method+"_embedding")
        adatas.uns['kBET_louvain_'+method] = index.kBET(adatas, batch_key="batch", label_key='louvain_'+method, type_="embed", embed=method+"_embedding")
        adatas.uns['kBET_mclust_'+method] = index.kBET(adatas, batch_key="batch", label_key='mclust_'+method, type_="embed", embed=method+"_embedding")
        print(adatas.uns['kBET_celltype'+method])
        print(adatas.uns['kBET_leiden_'+method])
        print(adatas.uns['kBET_louvain_'+method])
        print(adatas.uns['kBET_mclust_'+method])

    ##### PCR
    print('PCR:')
    for method in my_methods:
        print(method)
        adatas.uns['PCR'+method] = index.PCR(adata_pre, adatas, covariate="batch", embed=method+"_embedding", n_comps=min(adatas.obsm[method+'_embedding'].shape[1], 100))
        print(adatas.uns['PCR'+method])
    
    adatas.write('G:/dataset/2_saggital/output/harmony_K'+str(k)+'_result.h5ad')
```
