```python
from scipy.spatial.distance import cosine,correlation,euclidean
from scipy.spatial import distance

method = 'jaccard'
dis_list = []
for i1 in range(len(samples)):
    sample_dis_list = []
    for i2 in range(len(samples)):
        temp_dis = []
        for j in range(20):
            if method == 'cosine':
                dis = distance.cosine(adata_list[i1].uns['X_pca'][:,j], adata_list[i2].uns['X_pca'][:,j])
            if method == 'correlation':
                dis = distance.correlation(adata_list[i1].uns['X_pca'][:,j], adata_list[i2].uns['X_pca'][:,j])
            if method == 'euclidean':
                dis = distance.euclidean(adata_list[i1].uns['X_pca'][:,j], adata_list[i2].uns['X_pca'][:,j])
            if method == 'jaccard':
                dis = distance.jaccard(adata_list[i1].uns['X_pca'][:,j], adata_list[i2].uns['X_pca'][:,j])
            temp_dis.append(dis)
            
        sample_dis_list.append(np.mean(temp_dis))
    dis_list.append(sample_dis_list)
        
dis_list   
df = pd.DataFrame(dis_list, index = samples, columns = samples)
print(df)
df.to_csv('G:/dataset/result_pic/data_batch/MOSTA_PCA_'+method+'_batch.csv')
```

```python
import matplotlib
import seaborn as sns

def get_colors(num):
    if num < 10:
        return sns.color_palette(palette='tab10', n_colors=num).as_hex()
    elif num < 20:
        return sns.color_palette(palette='tab20', n_colors=num).as_hex()
    else:
        return list(matplotlib.colors.cnames.values())[:num]
    
def get_markers(num):
    markers = ['o', 's', 'p', '*', 'H', 'D', 'P', 'X', (6, 2, 0), 'v', '^', '<', '>', '8', '1', '2', '3', '4', 'd', 'h', '+', (5, 1, 90), '.', (6, 2, 90), 'x', (7, 2, 0)]
    res = []
    while(num > len(markers)):
        res.extend(markers)
        num = num - len(markers)
        
    res.extend(markers[:num])
    return res
```

```python
def plot_batch_celltype_center(save_path = None, center_mode = 'mean', visilization_method = 'umap'):
    annotation = 'annotation'
    spatial_location = {}
    
    if annotation != None:
        for celltype in adata.obs[annotation].unique().tolist():
            for batch in adata.obs['batch'].unique().tolist():
                if visilization_method in ['umap', 'UMAP']:
                    spatial_location[celltype + '-' + batch] = adata.obsm['X_umap'][(adata.obs[annotation]==celltype) & (adata.obs['batch']==batch)]
                else:
                    spatial_location[celltype + '-' + batch] = adata.obsm['X_tsne'][(adata.obs[annotation]==celltype) & (adata.obs['batch']==batch)]
    else:
        for batch in adata.obs['batch'].unique().tolist():
            if visilization_method in ['umap', 'UMAP']:
                spatial_location[batch] = adata.obsm['X_umap'][adata.obs['batch']==batch]
            else:
                spatial_location[batch] = adata.obsm['X_tsne'][adata.obs['batch']==batch]
    # print(spatial_location)

    batchs =  sorted(adata.obs['batch'].unique().tolist())
    batch_marker = get_markers(len(batchs))
    marker_dict = dict(zip(batchs, batch_marker))
    if annotation != None:
        celltypes = sorted(adata.obs[annotation].unique().tolist())
        label_color = get_colors(len(celltypes))
        label_dict = dict(zip(celltypes, label_color))
    else:
        batch_marker = get_colors(len(batchs))
        marker_dict = dict(zip(batchs, batch_marker))
    

    import matplotlib.pyplot as plt
    fig = plt.figure(figsize = (9, 9))
    ax = plt.axes([0, 0.1, 0.8, 0.8])

    for b in batchs:
        if annotation != None:
            for c in celltypes:
                points = spatial_location[c + '-' + b]
                if center_mode == 'mean':
                    center = np.mean(points, axis=0)
                else:
                    center = np.median(points, axis=0)
                from scipy.spatial import distance
                dis = np.mean( [distance.euclidean(point, center) for point in points])
                ax.scatter(center[0], center[1],alpha = 1,c = label_dict[c], marker = marker_dict[b])
                if not np.isnan(dis):
                    ax.scatter(center[0], center[1],alpha = 0.3,c = label_dict[c], s = dis*300)
        else:
            points = spatial_location[b]
            if center_mode == 'mean':
                center = np.mean(points, axis=0)
            else:
                center = np.median(points, axis=0)
            from scipy.spatial import distance
            dis = np.mean( [distance.euclidean(point, center) for point in points])
            ax.scatter(center[0], center[1],alpha = 1,c = marker_dict[b])
            if not np.isnan(dis):
                ax.scatter(center[0], center[1],alpha = 0.3,c = marker_dict[b], s = dis*300)

    if annotation != None:
        patch_list = []
        for c in celltypes:
            import matplotlib.patches as mpatches
            patch_list.append( mpatches.Patch(color=label_dict[c], label=c) )
        patch_legend = ax.legend(handles=patch_list,bbox_to_anchor=(1.27, 0.6), loc=7, borderaxespad=0.)
        ax.add_artist(patch_legend)

    scatter_list = []
    for b in batchs:
        if annotation != None:
            scatt = ax.scatter([], [], c = 'black', marker = marker_dict[b], label=b)
        else:
            scatt = ax.scatter([], [], c = marker_dict[b], label=b)
        scatter_list.append( scatt )
    scatter_legend = ax.legend(handles=scatter_list,bbox_to_anchor=(1.2, 0.15), loc=7, borderaxespad=0.)

    plt.axis('off')
    plt.xlim((-7, 20))
    plt.ylim((-8, 22))
    ax.grid(False)

    if save_path is None:
        plt.show()
    else:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        
plot_batch_celltype_center()
```

```python
def plot_batch_celltype(save_path = None, visilization_method = 'umap'):
    import matplotlib.pyplot as plt
    import scanpy as sc
    annotation = 'annotation'
    if visilization_method == 'umap':
        if annotation != None:
            sc.pl.umap(adata, color=['batch', annotation], ncols=2,
                           wspace=0.5, show=False, legend_fontsize=12, legend_fontoutline=2, frameon=False)
        else:
            sc.pl.umap(adata, color=['batch'], ncols=1, wspace=0.5, show=False, legend_fontsize=12, legend_fontoutline=2, frameon=False)
    else:
        if annotation != None:
            sc.pl.tsne(adata, color=['batch', annotation], ncols=2,
                           wspace=0.5, show=False, legend_fontsize=12, legend_fontoutline=2, frameon=False)
        else:
            sc.pl.umap(adata, color=['batch'], ncols=1, wspace=0.5, show=False, legend_fontsize=12, legend_fontoutline=2, frameon=False)
    
    if save_path is not None:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')

plot_batch_celltype()
```

```python
def plot_celltype_individual(save_dir = None, visilization_method = 'umap'):
    import matplotlib.pyplot as plt
    import scanpy as sc
    
    annotation = 'annotation'
    celltypes = sorted(adata.obs[annotation].unique().tolist())
    
    for c in celltypes:
        if visilization_method == 'umap':
            sc.pl.umap(adata[adata.obs[annotation]==c], color=['batch'], ncols=1, title = c,
                           wspace=0.5, show=False, legend_fontsize=12, legend_fontoutline=2, frameon=False)
        else:
            sc.pl.tsne(adata[adata.obs[annotation]==c], color=['batch'], ncols=1, title = c,
                           wspace=0.5, show=False, legend_fontsize=12, legend_fontoutline=2, frameon=False)
        if save_dir is not None:
            plt.savefig(save_dir + c.replace('/', '-') + '.png', dpi=300, bbox_inches='tight')
        
plot_celltype_individual()
```

```python

def cal_dis_min(df, celltype, batch):
    series_row = df.loc[celltype+'-'+batch, :]  #取行
    series_col = df.loc[:, celltype+'-'+batch]   #取列
    dis_min_batch = []
    dis_min_celltype = []
    for key, value in series_row.items():
        c, b = key.split('-', 1)
        if c == celltype and b != batch:
            dis_min_celltype.append(value)
        if b == batch and c != celltype:
            dis_min_batch.append(value)

    for key, value in series_col.items():
        c, b = key.split('-', 1)
        if c == celltype and b != batch:
            dis_min_celltype.append(value)
        if b == batch and c != celltype:
            dis_min_batch.append(value)

    # print(dis_min_celltype)
    # print(dis_min_batch)
    # print('dis_min_batch: ', np.nanmean(dis_min_batch))
    # print('dis_min_celltype: ', np.nanmean(dis_min_celltype))
    return np.nanmean(dis_min_batch), np.nanmean(dis_min_celltype)

res = []
for celltype in celltypes:
    for batch in batchs:
        try:
#             print(celltype, batch)
#             print(cal_dis_min(df_dis_min, celltype, batch))
            res.append(cal_dis_min(df_dis_min, celltype, batch))
        except Exception as e:
            print(celltype, batch, 'error ', e)

            
print(res)
print(np.nanmean(res, axis=0))
```

```python
def cal_dis_center(df, celltype, batch):
    target_df = df[(df['batch'] == batch) & (df['celltype'] == celltype)]
    assert target_df.shape[0]==1,print('check target_df')
    point = [target_df['x'].values[0], target_df['y'].values[0]]
    # print(point)
    assert not any(np.isnan(point)),print('point not exist')

    dis_center_batch = []
    for index, row in df[df['batch'] == batch].iterrows():
        if row['celltype'] not in [celltype, 'na']:
            dis = distance.euclidean(point, [row['x'], row['y']])
            # print(dis)
            dis_center_batch.append(dis)

    dis_center_celltype = []
    for index, row in df[df['celltype'] == celltype].iterrows():
        if np.isnan(row['x']) or np.isnan(row['y']):
            continue
            # print('batch:', row['batch'], ' celltype:', celltype, ' not exist')
        elif row['batch'] == batch:
            # print('skip itself')
            continue
        else:
            dis = distance.euclidean(point, [row['x'], row['y']])
            # print(dis)
            dis_center_celltype.append(dis)

    # print('dis_center_batch: ', np.nanmean(dis_center_batch))
    # print('dis_center_celltype: ', np.nanmean(dis_center_celltype))
    return np.nanmean(dis_center_batch), np.nanmean(dis_center_celltype)

res = []
for celltype in celltypes:
    for batch in batchs:
        try:
#             print(celltype, batch)
#             print(cal_dis_center(df_center, celltype, batch))
            res.append(cal_dis_center(df_center, celltype, batch))
        except Exception as e:
            print(celltype, batch, 'error ', e)

print(res)
print(np.nanmean(res, axis=0))
```
